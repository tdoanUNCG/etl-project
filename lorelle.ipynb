```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# import and preview investment csv
investments_df = pd.read_csv("resources/investments.csv", delimiter=",", encoding = 'unicode_escape')
investments_df
```

```python
# import and preview income csv
income_df = pd.read_csv("resources/kaggle_income.csv", delimiter=",", encoding = 'unicode_escape')
income_df.head()
```

```python
# clean investments csv
investments_df.columns
investments_df = investments_df[['name', 'homepage_url', ' market ',
       ' funding_total_usd ', 'status', 'country_code', 'state_code', 'region',
       'city', 'funding_rounds', 'founded_at', 'founded_month',
       'founded_quarter', 'founded_year', 'first_funding_at',
       'last_funding_at', 'seed', 'venture', 'equity_crowdfunding',
       'undisclosed', 'convertible_note', 'debt_financing', 'angel', 'grant',
       'private_equity', 'post_ipo_equity', 'post_ipo_debt',
       'secondary_market', 'product_crowdfunding']]
investments_df

# drop null values
investments_df = investments_df.dropna()
investments_df

#investments_df.dtypes

```

```python
# convert funding_total_usd to float
investments_df[' funding_total_usd '] = investments_df[' funding_total_usd '].astype("str")
investments_df[' funding_total_usd '] = investments_df[' funding_total_usd '].str.replace(',','').str.replace('-','0')
investments_df[' funding_total_usd '] = pd.to_numeric(investments_df[' funding_total_usd '], errors='coerce')
investments_df
```

```python
# filter for USA only
investments_df = investments_df.loc[investments_df['country_code']=='USA']
investments_df
```

```python
# clean income df - remove extraneous columns
income_df = income_df[['State_Name', 'State_ab','Mean', 'Median', 'Stdev', 'sum_w']]
income_df
```

```python
# group by and summarize by state name
groupby_state = income_df.groupby(['State_ab'])
median_income = groupby_state['Median'].mean()
mean_income = groupby_state['Mean'].mean()
stdev_income = groupby_state['Stdev'].mean()

# create new dataframe with values above
income_df_grouped = pd.DataFrame({"Median Income": median_income,
                                 "Mean Income": mean_income,
                                 "Standard Deviation": stdev_income})
# reset index
income_df_grouped.reset_index()

# export as csv
income_df_grouped.to_csv('resources/state.csv')
```

```python
# join both dfs by State_ab and state_code
investment_income = pd.merge(investments_df, income_df_grouped, how='left', left_on='state_code', right_on= 'State_ab')
investment_income
```

```python
# groupby  status
groupby_status = investment_income.groupby(['status'])
funding = groupby_status[' funding_total_usd '].mean()
status_funding = pd.DataFrame({"Funding Mean": funding})
status_funding
```

```python
# groupby state
groupby_state = investment_income.groupby(['state_code'])
status_state = groupby_state['status'].value_counts()
status_state = pd.DataFrame({'Status':status_state})
status_state
```

```python
# range of funding 
max_funding = investment_income[' funding_total_usd '].max()
print(f'Max investment is {max_funding}')

min_funding = investment_income[' funding_total_usd '].min()
print(f'Min investment is {min_funding}')

mean_funding = investment_income[' funding_total_usd '].mean()
print(f'Mean investment is {mean_funding}')

median_funding = investment_income[' funding_total_usd '].median()
print(f'Median investment is {median_funding}')
```

```python
# create bins for FUNDING
income_bins = [0, 500000, 1000000, 2000000, 6000000 , 50000000, 100000000, 500000000, 31000000000]
group_names = ["< 50,000","50,000 - 1 million", "1 - 2 million", "2 - 6 million", "6 - 50 million","50 - 100 million", "100 - 500 million", "> 500 million"]
```

```python
# create new column on investment ranges and set it as index
investment_income["Funding Range"]= pd.cut(investment_income[' funding_total_usd '],income_bins,labels=group_names).values.add_categories('0')
investment_income
```

```python
# groupby school size groupings
investment_groups = investment_income.groupby(['Funding Range'])
investmentrange = investment_groups['Funding Range'].count()
investmentrange
```

```python
# create df
investment_df = pd.DataFrame({"Number of companies": investmentrange})
investment_df.reset_index()
```

```python
# export to csv
investment_df.to_csv("resources/fundingrange.csv")
```

```python
#market by mean funding
market_by_mean_funding = round(investment_income.groupby(" market ")[" funding_total_usd "].mean(), 2)
pd.DataFrame(market_by_mean_funding)
pd.DataFrame(market_by_mean_funding.sort_values(ascending=False, na_position="last"))

# status df
status_df = investment_income.iloc[:,[0,4]]
status_df

# export to csv
status_df.to_csv("resources/status.csv", index=False)
```

```python
# company df
company_df = investment_income.loc[:,["name","state_code"," funding_total_usd ","homepage_url"," market ", "Funding Range"]]
company_df

#export to csv
company_df.to_csv("resources/company.csv", index=False)

company_df
```
