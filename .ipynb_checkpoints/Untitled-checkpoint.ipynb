```python
import pandas as pd
import csv
```

```python
investment = pd.read_csv("resources/investments.csv", delimiter=",", encoding = "unicode_escape")
investment.head()
```

```python
income = pd.read_csv("resources/kaggle_income.csv", delimiter=",", encoding = "unicode_escape")
income.head()
```

```python
#droping unessesary columns for analysis 
investment_df = investment.drop(['round_A', 'round_B', 'round_C', 'round_D', 'round_E', 'round_F', 'round_G', 'round_H'], axis=1)
income_df = income.drop(['State_Name', 'State_Code', 'Place', 'Area_Code', 'Primary', 'ALand', 'AWater', 'id', 'City', 'Lat', 'Lon', 'sum_w', 'County'], axis=1)
income_df.head()
```

```python
#looking only at USA, drop NaN state_codes, and remove weird commas
usa_investments = investment_df.loc[investment_df["country_code"] == "USA", :]
usa_investments.dropna(subset=['state_code'])
usa_investments[" funding_total_usd "] = usa_investments[" funding_total_usd "].str.replace(",","")
usa_investments.head()
```

```python
#preparing income df for merge
new_income = round(income_df.groupby("State_ab")["Mean"].mean(), 2)
pd.DataFrame(new_income).head()
```

```python
#merging df on state ab
usamerge_investments = usa_investments.merge(new_income, left_on='state_code', right_on='State_ab', how='inner')
```

```python
usamerge_investments.to_csv('merged_inv_income.csv')
```
